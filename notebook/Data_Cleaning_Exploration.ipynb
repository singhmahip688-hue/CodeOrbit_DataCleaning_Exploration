```python
# ==========================================
# Data Cleaning & Exploration with Pandas
# CodeOrbit Internship Task
# ==========================================

import pandas as pd
import numpy as np

# -----------------------------
# Load Dataset
# -----------------------------

df = pd.read_csv(r"C:\Users\singh\Downloads\employee_data.csv")

print("Original Dataset\n")
print(df)

print("\nDataset Shape:")
print(df.shape)

# -----------------------------
# Dataset Information
# -----------------------------

print("\nDataset Info")
print(df.info())

# -----------------------------
# Missing Values
# -----------------------------

print("\nMissing Values")
print(df.isnull().sum())

# -----------------------------
# Remove Duplicate Rows
# -----------------------------

duplicates = df.duplicated().sum()

print("\nDuplicate Rows:", duplicates)

df.drop_duplicates(inplace=True)

# -----------------------------
# Correct Data Types
# -----------------------------

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
df["Experience"] = pd.to_numeric(df["Experience"], errors="coerce")

# -----------------------------
# Fill Missing Values
# -----------------------------

df["Age"].fillna(df["Age"].median(), inplace=True)

df["Salary"].fillna(df["Salary"].mean(), inplace=True)

df["Experience"].fillna(df["Experience"].median(), inplace=True)

df["Department"].fillna("Unknown", inplace=True)

# -----------------------------
# Check Again
# -----------------------------

print("\nMissing Values After Cleaning")

print(df.isnull().sum())

# -----------------------------
# Basic Statistics
# -----------------------------

print("\nBasic Statistics")

print(df.describe())

# -----------------------------
# Mean
# -----------------------------

print("\nMean Salary")

print(df["Salary"].mean())

print("\nMean Age")

print(df["Age"].mean())

# -----------------------------
# Median
# -----------------------------

print("\nMedian Salary")

print(df["Salary"].median())

print("\nMedian Age")

print(df["Age"].median())

# -----------------------------
# Value Counts
# -----------------------------

print("\nDepartment Counts")

print(df["Department"].value_counts())

print("\nGender Counts")

print(df["Gender"].value_counts())

# -----------------------------
# Highest Salary
# -----------------------------

highest = df.loc[df["Salary"].idxmax()]

print("\nEmployee with Highest Salary")

print(highest)

# -----------------------------
# Lowest Salary
# -----------------------------

lowest = df.loc[df["Salary"].idxmin()]

print("\nEmployee with Lowest Salary")

print(lowest)

# -----------------------------
# Save Clean Dataset
# -----------------------------

df.to_csv("clean_employee_data.csv", index=False)

print("\nClean Dataset Saved Successfully")
```

```python
# Bar Chart
import matplotlib.pyplot as plt

department = df["Department"].value_counts()

plt.figure(figsize=(7,5))

department.plot(kind="bar")

plt.title("Employees by Department")

plt.xlabel("Department")

plt.ylabel("Count")

plt.tight_layout()

plt.show()
```

```python
# Salary Histogram
plt.figure(figsize=(7,5))

plt.hist(df["Salary"], bins=5)

plt.title("Salary Distribution")

plt.xlabel("Salary")

plt.ylabel("Frequency")

plt.tight_layout()

plt.show()
```
